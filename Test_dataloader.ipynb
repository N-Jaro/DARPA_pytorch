```python
from torch.utils.data import Dataset, DataLoader
from libs.Dataset_val import PatchDataGenerator 
import matplotlib.pyplot as plt
import numpy as np
```

```python
generator = PatchDataGenerator(patch_size=256, norm_type='basic', hue_factor=1, augment=True)
```

```python
dataloader = DataLoader(generator, batch_size=5, shuffle=True, num_workers=1)
```

```python
for images, labels in dataloader:
    images = images
    labels = labels
    break
```

```python
images
```

```python
# Define the number of items from the batch to visualize
items_to_visualize = 5

plt.figure(figsize=(12, 3 * items_to_visualize))  # Adjust figure size dynamically

for idx in range(items_to_visualize):
    image = images[idx]
    label = labels[idx]

    plt.subplot(items_to_visualize, 3, idx * 3 + 1)
    plt.imshow(np.transpose(image[:3, ...], (1, 2, 0)))
    plt.title(f'Item {idx+1} - First 3 Channels')

    plt.subplot(items_to_visualize, 3, idx * 3 + 2)
    plt.imshow(np.transpose(image[3:, ...], (1, 2, 0)))
    plt.title(f'Item {idx+1} - Remaining Channels')

    plt.subplot(items_to_visualize, 3, idx * 3 + 3)
    plt.imshow(np.transpose(label, (1, 2, 0)))
    plt.title(f'Item {idx+1} - Label')

plt.show()
```
